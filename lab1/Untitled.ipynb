```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
%matplotlib inline
```

```python
tbl = pd.read_table('mpp.txt', sep="   ", engine="python", header=None)
```

```python
tbl.drop(tbl.columns[395:400],axis=1,inplace=True)
```

```python
mv = (tbl.max()).max()
```

```python
dfn = tbl / mv
```

```python
sns.set()
```

```python
ax = sns.heatmap(dfn)
```
